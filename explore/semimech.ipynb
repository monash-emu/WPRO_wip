```python
from typing import Dict
from scipy.stats import gamma
import numpy as np
import pandas as pd
```

```python
def get_gamma_params_from_mean_sd(req_mean: float, req_sd: float) -> Dict[str, float]:
    var = req_sd ** 2.0
    scale = var / req_mean
    a = req_mean / scale
    return {'a': a, 'scale': scale}
```

```python
req_sd = 2.0
req_mean = 4.0
gamma_params = get_gamma_params_from_mean_sd(req_mean, req_sd)
```

```python
x_vals = np.linspace(0.0, 10.0, 100)
gamma_pdf_vals = gamma.pdf(x_vals, **gamma_params)
```

```python
pd.Series(gamma_pdf_vals, index=x_vals).plot()
```
